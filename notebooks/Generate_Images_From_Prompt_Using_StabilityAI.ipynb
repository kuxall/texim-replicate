```python
from diffusers import DiffusionPipeline
import torch
```

```python
pipe = DiffusionPipeline.from_pretrained("stabilityai/stable-diffusion-xl-base-1.0", torch_dtype=torch.float16, variant="fp16")
pipe.to("cuda")
```

```python
prompt = "an old man carpenter building an intricate wooden toy boat with various gears and cogs, concentrated, fluffy mustache, workshop filled with christmas decorations, nighttime, moody, motivated lighting, practical lighting, sitting at a wooden table, open window behind him, cute, nostalgic, masterpiece"
image = pipe(prompt=prompt).images[0]
image
```

```python
from google.colab import files
output_path = "/content/generated_image.jpg"  # Change the path and filename as desired
image.save(output_path)

# Download the image
files.download(output_path)
```
